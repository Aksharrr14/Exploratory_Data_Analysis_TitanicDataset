```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#now we will load the titanic data set for exploratory data analysis- which involves examining dataset's structure,visualizng data,and identifying patterns
titanic = pd.read_csv('C:\\Users\\HP\\Downloads\\Titanic.csv')

# View first five rows of the dataset
titanic.head()
```

```python
titanic.shape
```

```python
titanic.isnull().sum()
```

```python
sb.countplot(x='Survived',data=titanic)
plt.show()
```

```python
titanic.groupby(['Sex', 'Survived'])['Survived'].count()
```

```python
titanic[['Sex','Survived']].groupby(['Sex']).mean().plot.bar()
sb.countplot(x='Sex',hue='Survived',data=titanic)
plt.show()
```

```python
sb.catplot(x='Survived', hue='Sex', data=titanic)
plt.show()
```

```python
print('Oldest person Survived was of:',titanic['Age'].max())
print('Youngest person Survived was of:',titanic['Age'].min())
print('Average person Survived was of:',titanic['Age'].mean())
```
